```python
from pylab import plot,show, ylim, xlabel, ylabel
y = [ 1.0, 2.4, 1.7, 0.3, 0.6, 1.8 ]
plot(y)
show()
```

```python
x = [ 0.5, 1.0, 2.0, 4.0, 7.0, 10.0 ]
y = [ 1.0, 2.4, 1.7, 0.3, 0.6, 1.8 ]
plot(x,y)
show()
```

```python
from numpy import linspace,sin
x= linspace(0,10,100)
y = sin(x)

plot(x,y)
ylim(-1.1, 1.1)
xlabel('x')
ylabel('sin(x)')
show()
```

```python
plot(x,y,"g--")
plot(x,y,"k.")
show()
```
